```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
Airline=pd.read_csv("EastWestAirlines.csv")
Airline
```

## Hierarchical Clustering

```python
# Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
## Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Airline.iloc[:,1:11])
```

```python
df_norm
```

```python
## Create Dendogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
## Create Clusters
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='single')
```

```python
## Save Clusters for chart
y_hc=hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Airline['h_clusterid']=Clusters
```

```python
Airline.sort_values('h_clusterid')
```

```python
### Group these clusters
Airline.groupby('h_clusterid').agg(['mean'])
```

### Passengers falling in cluster 1,2,3 are likely to be offered Award flights


## KMeans Clustering

```python
## Normalization Function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Airline_df = scaler.fit_transform(Airline.iloc[:,:11])
```

```python
scaled_Airline_df
```

```python
## How to find optimum numbers of clusters
## The K-means algorithm aims to choose centroid that minimize the inertia ,or within cluster sum of square criterion
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(scaled_Airline_df)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range (1,15):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_Airline_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## Build Cluster Algorithm
from sklearn.cluster import KMeans
clusters_new=KMeans(7,random_state=42)
clusters_new.fit(scaled_Airline_df)
```

```python
clusters_new.labels_
```

```python
## Assigning clusters to the dataset
Airline['clusterid_new'] = clusters_new.labels_
```

```python
## These are Standarized values
clusters_new.cluster_centers_
```

## DBSCAN

```python
stscaler = StandardScaler().fit(array)
x = stscaler.transform(array)
```

```python
x
```

```python
dbscan = DBSCAN(eps=0.8,min_samples=6)
dbscan.fit(x)
```

```python
## Noisy Samples are given the label -1
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```
