```python
### Checking for nul values 
crime_ds.isnull().any()
```

```python
### Normalization function
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
## Normalized dataframe(Considering the numerical part of the data only)
dF_norm=norm_func(crime_ds.iloc[:,1:])
```

```python
dF_norm
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='average')
## in affinity you can specify whichever distance you want to use
```

```python
# Save cluster for chart
y_hc=hc.fit_predict(dF_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
crime_ds['h_clusterid']= Clusters
```

```python
crime_ds.sort_values('h_clusterid')
```

```python
### Group these clusters
crime_ds.groupby('h_clusterid').agg(['mean'])
```

### Inferences
####  States in cluster 1 has minmum crime rate as compare to others because it has minimum mean. 
#### States in cluster 2 has max assult case as compare to other clusters .

##  K-means clustering

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
DS=pd.read_csv("crime_data.csv")
DS
```

```python
## Normalization function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df1= scaler.fit_transform(DS.iloc[:,1:])
```

```python
scaled_crime_df1
```

#### How to find optimum no of cluster
#### The KMeans algorithm aims to choose centroid that minimize the inertia,or within the clusters sum-of-squares criterion
The number of clusters has to be be decided when k-means clustering is used unlike hierarchical clustering. Let's start with creating scree plot. Scree plot is a plot between WCSS (Within cluster sum of squares) and number of clusters. Without sound domain knowledge or in the scenarios with unclear motives, the scree plots help us decide the number of clusters to specify.

### What is random state
To set a seed and make the randomization more deterministic and we can take any variable according to our call

```python
kmeans = KMeans(n_clusters=4,random_state=0)
kmeans.fit(scaled_crime_df1)
```

```python
kmeans.inertia_
```

```python
wcss= []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df1)
    wcss.append(kmeans.inertia_)
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df1)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(scaled_crime_df1)

y_kmeans
```

```python
#beginning of  the cluster numbering with 1 instead of 0
y_kmeans1=y_kmeans+1

# New list called cluster
cluster = list(y_kmeans1)
# Adding cluster to our data set
DS['cluster'] = cluster
```

```python
## Mean of cluster 1 to 4
kmeans_mean_cluster = pd.DataFrame(round(DS.groupby('cluster').mean(),1))
kmeans_mean_cluster
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.scatterplot(x=DS['Murder'], y = DS['Assault'],hue=y_kmeans1)
```

The above scatter shows a distribution of how the states are scattered and clusters are visible based on Murders and Assaults. And there is a positive correlation between occurrence of Murder and Assault in different states.

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### import .csv file and covert it to a DataFrame object
crime=pd.read_csv("crime_data.csv")
crime
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime['clusters']=dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clusters'],crime['UrbanPop'], c=dbscan.labels_)
```

### Inferences

#### States come in the cluster 2 have minimum crime rate as compare to other 2 clusters
#### As compare to other crimes states in cluster 2 have minimum murder crime .
